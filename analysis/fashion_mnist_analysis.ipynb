```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import fashion_mnist
```

```python
# Cargar los datos
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
# Imprimir las dimensiones de los datos
print(f"Train images shape: {train_images.shape}")
print(f"Train labels shape: {train_labels.shape}")
print(f"Test images shape: {test_images.shape}")
print(f"Test labels shape: {test_labels.shape}")
```

```python
# Definir nombres de clases
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", 
               "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]

# Contar la frecuencia de cada clase
unique, counts = np.unique(train_labels, return_counts=True)
class_distribution = dict(zip(class_names, counts))

# Graficar la distribución de clases
plt.figure(figsize=(10, 6))
sns.barplot(x=list(class_distribution.keys()), y=list(class_distribution.values()))
plt.title("Distribución de Clases en el Conjunto de Entrenamiento")
plt.xlabel("Clase")
plt.ylabel("Frecuencia")
plt.xticks(rotation=45)
plt.show()
```

```python
# Mostrar algunas imágenes de ejemplo de cada clase
plt.figure(figsize=(10, 10))
for i in range(10):
    plt.subplot(5, 5, i + 1)
    idx = np.where(train_labels == i)[0][0]
    plt.imshow(train_images[idx], cmap='gray')
    plt.title(class_names[i])
    plt.axis('off')
plt.show()
```
